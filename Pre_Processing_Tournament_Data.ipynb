Delete all lines where the "loser_rank" attribute is null, because there are few occasions and this can cause errors in the results

```python
dados.dropna (subset =['loser_rank'],inplace=True)
```

```python
dados
```

Delete all columns that there are null values

```python
dados.dropna(axis=1,how='all')
```


Change value of data in a column

```python
dados['loser_entry'].unique()
```

Replace null values ​​with x

```python
dados['loser_entry'].fillna(value='x',inplace=True)
```

```python
dados['loser_entry'].unique()
```

Alterar mais de um atributo, com condições diferentes

```python
dados['l_1stIn'].unique()
```

```python
dados ['l_1stWon'].unique()
```

```python
dados.fillna(value={'l_1stIn':0,'l_1stWon':dados['l_1stWon'].mean()},inplace=True)
```

```python
dados['l_1stIn'].unique()
```

```python
dados['l_1stWon'].unique()
```

Use the ffill method to project a data to the next record

```python
dados['minutes'][1800:1990]
```

```python
dados['minutes'].fillna(method='ffill',inplace=True)
```

```python
dados['minutes'][1975:1990]
```

See Scikit-learn about preprocessing. <br> Transforming in values.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
countries=dados['winner_ioc']
```

```python
countries
```

```python
lencoder=LabelEncoder()
countries_numbers =lencoder.fit_transform(countries)
countries_numbers
```
