## How to make a shift


Let's import apple stock data

https://finance.yahoo.com/quote/AAPL/history/

```python
import pandas as pd
```

```python

dados=pd.read_csv('AAPL.csv')
```

```python
dados
```

```python
dados['close_yesterday']=dados['Close'].shift(1)
```

```python
dados
```

What was done: We tossed the information from the previous day that was on the top line to the bottom line so that we can see how stock closed the day before and the next day on the same line.
Below, a new column with the percentage change from one day to the next.

```python
dados['altered %']=((dados['Close']/dados['close_yesterday'])-1)*100
```

```python
dados.head(20)
```

```python
dados.dropna(subset=['altered %'], inplace=True)
```
